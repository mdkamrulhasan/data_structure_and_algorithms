# Data Structures Demonstration (Python)

## Array


```python
from array import array
```

```python
# Array example using a Python list
scores = array('i', [78, 85, 92, 88, 76])
print("Initial scores:", scores)
```

```python
# A Python List can be also used as a dynamic array
#scores = [78, 85, 92, 88, 76]
```

```python

print("Initial scores:", scores)
print("First student's score:", scores[0])
print("Last student's score:", scores[-1])


scores[2] = 95
print("Updated scores:", scores)

print("All scores:")
for score in scores:
    print(score)
```

```python
scores.insert(1, 89)
print("Scores after insertion:", scores)
```

```python
scores.remove(76)
print("Scores after removal:", scores)
```

```python
len(scores)
```

```python
scores.insert(4, -9)
```

```python
len(scores)
```
